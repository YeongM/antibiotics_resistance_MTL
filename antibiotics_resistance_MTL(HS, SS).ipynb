```python
def binary_cross_entropy(y_true, y_pred):
    mask = tf.math.is_finite(y_true[:,0])
    y_true = tf.boolean_mask(y_true, mask)
    y_pred = tf.boolean_mask(y_pred, mask)
    
    y_true = tf.debugging.check_numerics(y_true, "NaN or Inf detected in y_true")
    y_pred = tf.debugging.check_numerics(y_pred, "NaN or Inf detected in y_pred")
    
    bce = tf.keras.losses.BinaryCrossentropy()
    
    return bce(y_true, y_pred)
```

```python
def performacne_model(Real_Test_y, y_pred, y_pred_prob):
    accuracy= accuracy_score(Real_Test_y, y_pred)
    precision = precision_score(Real_Test_y, y_pred)
    recall = recall_score(Real_Test_y, y_pred)
    f1 = f1_score(Real_Test_y, y_pred)
    auc = roc_auc_score(Real_Test_y, y_pred_prob)
    prc = average_precision_score(Real_Test_y, y_pred_prob)
    confusion = confusion_matrix(Real_Test_y, y_pred)
    
    return [accuracy, precision, recall, f1, auc, prc, confusion]
```

```python
# Standard Deep Learning code
class HardSharing_Layer3():
    def base_node(self, x, node1, drop_out_rate1, node2, drop_out_rate2):
        x = Dense(node1)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate1)(x)
        x = Dense(node2)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate2)(x)
        return x

    def build_tgc_branch(self, x, node3, drop_out_rate3):
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="tgc_output")(x)
        return x
    
    def build_m_branch(self, x, node3, drop_out_rate3):
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="m_output")(x)
        return x
    
    def build_fgc_branch(self, x, node3, drop_out_rate3):   
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="fgc_output")(x)
        return x
    
    def build_f_branch(self, x, node3, drop_out_rate3):
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="f_output")(x)
        return x
    
    def build_cp_branch(self, x, node3, drop_out_rate3):   
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="cp_output")(x)
        return x
    
    def build_c_branch(self, x, node3, drop_out_rate3):   
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="c_output")(x)
        return x
    
    def build_a_branch(self, x, node3, drop_out_rate3):   
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="a_output")(x)
        return x
    
    def build_ag_branch(self, x, node3, drop_out_rate3):   
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="ag_output")(x)
        return x
    
    def build_g_branch(self, x, node3, drop_out_rate3):   
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        x = Dense(1)(x)
        x = Activation("sigmoid", name="g_output")(x)
        return x
    
    def assemble_full_model(self, feature, node1=64, drop_out_rate1=0.1, node2=64, drop_out_rate2=0.1, node3=64, drop_out_rate3=0.1):
        input_shape = (feature,)
        inputs = Input(shape=input_shape)
        
        x = self.base_node(inputs, node1, drop_out_rate1, node2, drop_out_rate2)
        
        tgc_branch = self.build_tgc_branch(x, node3, drop_out_rate3)
        m_branch = self.build_m_branch(x, node3, drop_out_rate3)
        fgc_branch = self.build_fgc_branch(x, node3, drop_out_rate3)
        f_branch = self.build_f_branch(x, node3, drop_out_rate3)
        cp_branch = self.build_cp_branch(x, node3, drop_out_rate3)
        c_branch = self.build_c_branch(x, node3, drop_out_rate3)
        a_branch = self.build_a_branch(x, node3, drop_out_rate3)
        ag_branch = self.build_ag_branch(x, node3, drop_out_rate3)
        g_branch = self.build_g_branch(x, node3, drop_out_rate3)
        
        model = Model(inputs=inputs,
                     outputs = [tgc_branch, m_branch, fgc_branch, f_branch, cp_branch, c_branch, a_branch, ag_branch,g_branch],
                     name="anti_net")
        return model
    
def model2_builder(node1=64, drop_out_rate1=0.1, node2=64, drop_out_rate2=0.1, node3=64, drop_out_rate3=0.1):
    loss_dict ={'tgc_output': binary_cross_entropy,
            'm_output': binary_cross_entropy,
            'fgc_output': binary_cross_entropy,
            'f_output': binary_cross_entropy,
            'c_output': binary_cross_entropy,
            'cp_output': binary_cross_entropy,
            'a_output': binary_cross_entropy,
            'ag_output': binary_cross_entropy,
            'g_output': binary_cross_entropy}
    
    metrics_dict = {'tgc_output': binary_cross_entropy,
                'm_output': binary_cross_entropy,
                'fgc_output': binary_cross_entropy,
                'f_output': binary_cross_entropy,
                'c_output': binary_cross_entropy,
                'cp_output': binary_cross_entropy,
                'a_output': binary_cross_entropy,
                'ag_output': binary_cross_entropy,
                'g_output': binary_cross_entropy}

    model = HardSharing_Layer3().assemble_full_model(feature_size, node1, drop_out_rate1, node2, drop_out_rate2, node3, drop_out_rate3)

    model.compile(optimizer='adam',
                  loss=loss_dict,
                  metrics=metrics_dict)

    return model
```

```python
class SoftSharing_Layer3():
    def base_node(self, x, node1, drop_out_rate1):
        x = Dense(node1)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate1)(x)
        return x
    
    def hidden_branch1(self,x, node2, drop_out_rate2):
        x = Dense(node2)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate2)(x)
        return x
    
    def hidden_branch2(self,x, node3, drop_out_rate3):
        x = Dense(node3)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(drop_out_rate3)(x)
        return x

    def build_tgc_branch(self, x):
        x = Dense(1)(x)
        x = Activation("sigmoid", name="tgc_output")(x)
        return x

    def build_m_branch(self, x):
        x = Dense(1)(x)
        x = Activation("sigmoid", name="m_output")(x)
        return x

    def build_fgc_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="fgc_output")(x)
        return x
    
    def build_f_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="f_output")(x)
        return x
    
    def build_cp_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="cp_output")(x)
        return x
    
    def build_c_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="c_output")(x)
        return x
    
    def build_a_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="a_output")(x)
        return x
    
    def build_ag_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="ag_output")(x)
        return x
    
    def build_g_branch(self, x):   
        x = Dense(1)(x)
        x = Activation("sigmoid", name="g_output")(x)
        return x
    
    def assemble_full_model(self, feature, node1=64, drop_out_rate1=0.1, node2=64, drop_out_rate2=0.1, node3=64, drop_out_rate3=0.1):
        input_shape = (feature,)
        inputs = Input(shape=input_shape)
        
        tgc_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        m_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        fgc_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        f_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        cp_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        c_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        a_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        ag_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        g_branch1 = self.base_node(inputs, node1, drop_out_rate1)
        
        tgc_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        m_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        fgc_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        f_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        cp_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        c_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        a_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        ag_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        g_branch_tmp = Concatenate()([tgc_branch1,m_branch1,fgc_branch1,f_branch1,cp_branch1,c_branch1,a_branch1,ag_branch1,g_branch1])
        
        tgc_branch2 = self.hidden_branch1(tgc_branch_tmp, node2, drop_out_rate2)
        m_branch2 = self.hidden_branch1(m_branch_tmp, node2, drop_out_rate2)
        fgc_branch2 = self.hidden_branch1(fgc_branch_tmp, node2, drop_out_rate2)
        f_branch2 = self.hidden_branch1(f_branch_tmp, node2, drop_out_rate2)
        cp_branch2 = self.hidden_branch1(cp_branch_tmp, node2, drop_out_rate2)
        c_branch2 = self.hidden_branch1(c_branch_tmp, node2, drop_out_rate2)
        a_branch2 = self.hidden_branch1(a_branch_tmp, node2, drop_out_rate2)
        ag_branch2 = self.hidden_branch1(ag_branch_tmp, node2, drop_out_rate2)
        g_branch2 = self.hidden_branch1(g_branch_tmp, node2, drop_out_rate2)
        
        tgc_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        m_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        fgc_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        f_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        cp_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        c_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        a_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        ag_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        g_branch_tmp = Concatenate()([tgc_branch2,m_branch2,fgc_branch2,f_branch2,cp_branch2,c_branch2,a_branch2,ag_branch2,g_branch2])
        
        tgc_branch3 = self.hidden_branch2(tgc_branch_tmp, node3, drop_out_rate3)
        m_branch3 = self.hidden_branch2(m_branch_tmp, node3, drop_out_rate3)
        fgc_branch3 = self.hidden_branch2(fgc_branch_tmp, node3, drop_out_rate3)
        f_branch3 = self.hidden_branch2(f_branch_tmp, node3, drop_out_rate3)
        cp_branch3 = self.hidden_branch2(cp_branch_tmp, node3, drop_out_rate3)
        c_branch3 = self.hidden_branch2(c_branch_tmp, node3, drop_out_rate3)
        a_branch3 = self.hidden_branch2(a_branch_tmp, node3, drop_out_rate3)
        ag_branch3 = self.hidden_branch2(ag_branch_tmp, node3, drop_out_rate3)
        g_branch3 = self.hidden_branch2(g_branch_tmp, node3, drop_out_rate3)
        
        tgc_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        m_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        fgc_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        f_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        cp_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        c_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        a_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        ag_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        g_branch_tmp = Concatenate()([tgc_branch3,m_branch3,fgc_branch3,f_branch3,cp_branch3,c_branch3,a_branch3,ag_branch3,g_branch3])
        
        tgc_branch = self.build_tgc_branch(tgc_branch_tmp)
        m_branch = self.build_m_branch(m_branch_tmp)
        fgc_branch = self.build_fgc_branch(fgc_branch_tmp)
        f_branch = self.build_f_branch(f_branch_tmp)
        cp_branch = self.build_cp_branch(cp_branch_tmp)
        c_branch = self.build_c_branch(c_branch_tmp)
        a_branch = self.build_a_branch(a_branch_tmp)
        ag_branch = self.build_ag_branch(ag_branch_tmp)
        g_branch = self.build_g_branch(g_branch_tmp)
        
        model = Model(inputs=inputs,
                     outputs = [tgc_branch, m_branch, fgc_branch, f_branch, cp_branch, c_branch, a_branch, ag_branch, g_branch],
                     name="anti_net")
        return model
    
def model2_builder(node1=64, drop_out_rate1=0.1, node2=64, drop_out_rate2=0.1, node3=64, drop_out_rate3=0.1):
    loss_dict ={'tgc_output': binary_cross_entropy,
            'm_output': binary_cross_entropy,
            'fgc_output': binary_cross_entropy,
            'f_output': binary_cross_entropy,
            'c_output': binary_cross_entropy,
            'cp_output': binary_cross_entropy,
            'a_output': binary_cross_entropy,
            'ag_output': binary_cross_entropy,
            'g_output': binary_cross_entropy}

    metrics_dict = {'tgc_output': binary_cross_entropy,
                'm_output': binary_cross_entropy,
                'fgc_output': binary_cross_entropy,
                'f_output': binary_cross_entropy,
                'c_output': binary_cross_entropy,
                'cp_output': binary_cross_entropy,
                'a_output': binary_cross_entropy,
                'ag_output': binary_cross_entropy,
                'g_output': binary_cross_entropy}

    model = SoftSharing_Layer3().assemble_full_model(feature_size, node1, drop_out_rate1, node2, drop_out_rate2, node3, drop_out_rate3)

    model.compile(optimizer='adam',
                  loss=loss_dict,
                  metrics=metrics_dict)

    return model
```
